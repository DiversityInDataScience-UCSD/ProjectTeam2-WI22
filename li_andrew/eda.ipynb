```python
# plot race by timeserv, totdays. comdum, 

cols = ['TIMESERV', 'TOTDAYS', 'COMDUM', 'AGE']
data['TIMESERV'] = pd.to_numeric(data['TIMESERV'], errors = 'coerce')#['TIMESERV', 'TOTDAYS', 'COMDUM', 'AGE']
data['AGE'] = pd.to_numeric(data['AGE'], errors = 'coerce')

for i in cols:
    sns.catplot(x = 'NEWRACE', y = i, data = data)
    plt.show()
```
